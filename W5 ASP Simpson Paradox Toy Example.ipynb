### Example
- if Tux is a penguin, can Tux fly? 
- if Tweety is a chicken, can Tweety fly?

#### Negation is used in this example. 
It will generate two possible worlds
- one world tux can fly
- one world tux can not fly

```python
%%clingo --run --donot-display_input
% facts
bird(tux). penguin(tux).
bird(tweety). chicken(tweety).
% rules
flies(X) :- bird(X), not -flies(X).
-flies(X) :- bird(X), not flies(X).
-flies(X) :- penguin(X).
% constrain
:- flies(tux), -flies(tux).
:- flies(tweety), -flies(tweety).
#show flies/1.
```

```python
%%clingo --run  --donot-display_input
peg(a;b;c).
disk(1..4).
init_on(1..4,a).
goal_on(1..4,c).
moves(15).
%Generate
1{ move(D,P,T): disk(D): peg(P) }1 :- moves(M), T = 1..M.
%Define
move(D,T) :- move(D,_,T).
on(D,P,0) :- init_on(D,P).
on(D,P,T) :- move(D,P,T).
on(D,P,T+1):- on(D,P,T), not move(D,T+1), not moves(T).
blocked(D-1,P,T+1) :- on(D, P, T), disk(D), not moves(T).
blcoked(D-1,P,T)   :- blocked(D,P,T), disk(D).
%Test
:- move(D,P,T), blocked(D-1,P,T).
:- move(D,T), on(D,P,T-1), blocked(D,P,T).
:- goal_on(D,P), not on(D,P,M), moves(M).
:- not 1{on(D,P,T):peg(P)}1, disk(D), moves(M),T=1..M.
#hide.
#show move/3.

```

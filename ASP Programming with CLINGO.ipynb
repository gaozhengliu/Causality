# 5.0 Combinatorial Search

```python
%%clingo --run --donot-display_input
% Seat Arrangement

% For each chair, there is exactly one guest
1{at(1..n, C)} 1 :- C = 1..n.

% No guest sit on two chairs.
C1 = C2 :- at(G,C1), at(G,C2).

% Two chairs are next to each other
adj(X,Y) : X=1..n, Y=1..n, |X-Y| = 1.
adj(1, n; n, 1).

:- like(G1,G2), at(G1,C1), at(G2,C2), not adj(C1,C2).
:- dislike(G1,G2), at(G1,C1), at(G2,C2), adj(C1,C2).

#show at/2.

#const n = 6.
like(1,2; 3,4).
dislke(2,3; 1,3).
```

### Schur Numbers
A set X of numbers is called sum-free if the sum of two elements of X never belongs to X. Can we partition the set {1,...,n} into 2 sum-free subsets?  This is possible if n = 4: both {1,4} and {2,3} are sum-free. But if n = 5 then such a partition doesn’t exist.

```python
%%clingo --run --donot-display_input


% Partition {1, ..., n} into r sum-free sets
#const r = 3.
#const n = 9.

% achieved: Set{1,...,n} is partitioned into subsets {I: in(I,1), ... , {I: in(I,r)}}

1{in(I, 1..r)}1 :- I = 1..n.

% achieved: these subsets are sum free
:- in(I,X), in(J,X), in(I+J,X).
```
